```python
#gemma_lm = keras_nlp.models.GemmaCausalLM.from_preset("gemma_instruct_2b_en")
gemma_lm = keras_nlp.models.GemmaCausalLM.from_preset("Gemma")
gemma_lm.generate("Keras is a", max_length=30)

# Generate with batched prompts.
gemma_lm.generate(["Keras is a", "I want to say"], max_length=30)
```

```python
from datetime import datetime

t1 = datetime.now()
m = gemma_lm.generate("今日は寒いね。風も強いし。どんな格好で外に出たらいいかな？", max_length=256)
t2 = datetime.now()
print(m)
print(t2-t1)
```
